# 第2章 順伝播型ネットワーク

## 2.2 活性化関数
基本的な活性化関数と，その他の活性化関数の例をいくつか紹介

```python
# シグモイド関数

## ロジスティックシグモイド関数（a.k.a ロジスティック関数）
def logistic(u):
    return 1 / (1 + sp.exp(-u))

## 双曲線正接関数
def tanh(u):
    return sp.tanh(u)

# 正則化線形関数
def rectifier(u):
    return max(u, 0)
```

```python
X = np.linspace(-5.0, 5.0, num=100)

plt.figure(figsize=(11, 7))

plt.plot(X, np.vectorize(logistic, otypes=[np.float])(X), 'r', label='logistic')
plt.plot(X, np.vectorize(tanh, otypes=[np.float])(X), 'g--', label='tanh')
plt.plot(X, np.vectorize(rectifier, otypes=[np.float])(X), 'b.', label='rectifier')

plt.ylim(ymin=-1.5, ymax=1.5)
plt.title('Activation Functions (Basic)')
plt.legend(loc='upper left')
```

```python
# 線形写像，恒等写像
def linear(u):
    return u

# ロジスティック関数を区分的に直線で近似
def linear_like_logistic(u):
    if u < -1.: return -1
    elif u >= 1.: return 1
    else: return u
```

```python
plt.figure(figsize=(11, 7))

plt.plot(X, np.vectorize(linear, otypes=[np.float])(X), 'r', label='linear')
plt.plot(X, np.vectorize(linear_like_logistic, otypes=[np.float])(X), 'g--', label='linear (like logistic)')

plt.ylim(ymin=-1.5, ymax=1.5)
plt.title('Activation Functions (Others)')
plt.legend(loc='upper left')
```
